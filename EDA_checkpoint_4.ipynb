```
##Most common words in descriptions
def remove_stopwords(token):
    mt_stopwords = nltk.corpus.stopwords.words("english")
    mt_stopwords.extend(['wine','drink'])
    return token.lower() not in mt_stopwords and token.isalnum() 

def lemm(text):
    text = wordnet_lemmatizer.lemmatize(text)
    return text

def cleaning(text):
    # text is str
    #text = re.sub(pattern2, '', text) # removes @ppl
    #text = re.sub('#\w+', '', text) # removes '#'
    text = re.sub('htt[\w.-/]+', '', text) # remove urls
    text = re.sub("n't", ' not', text)
    text = re.sub("arenot", 'are not', text)
    # turn into list
    tokens = nltk.word_tokenize(text.lower())
    #tokens_lower = list(map(lambda t: t.lower(), tokens))
    tokens = list(filter(remove_stopwords, tokens))  
    tokens = list(map(lemm, tokens))
    #print(tokens)
    return " ".join(tokens)

def most_common(sentence):
    toke = nltk.word_tokenize(sentence)
    return dict(nltk.FreqDist(toke).most_common(30))
```

```
df_total_nlp['text_clean']=df_total_nlp['description'].apply(cleaning)
new = ' '.join(df_total_nlp['text_clean'])
new

freq_dis = most_common(new)

wordcloud = WordCloud(width=480, height=480, max_font_size=200, min_font_size=10)

dico = nltk.FreqDist(freq_dis)
wordcloud.generate_from_frequencies(dico)

plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```
